```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Cargamos los datos desde la pagina del profe

```python
url = "https://dmontero.org/mineria/datasets/Churn_Modelling.csv"
df = pd.read_csv(url)
df.head(10)
```

Los datos vienen limpios y sin nulos
Se quiere predecir cuales son los clientes que se van a ir de la empresa/ La feature que se quiere predecir es la que esta en la ultima columna
No vamos a seleccionar las 3 primeras columnas, porque no aportan al analisis.


```python
# se deja la feature Exited separada en una variable
# Es una convención llamar a esta varaible y.
y = df['Exited']
y.head()
```

```python
# Por ahora dejo los datos que me interesan ignorando las 3 primeras columnas. Por convencion a la variable resultante se le llama X 
x = df.iloc[:, 3: -1]
x.head()
```

```python
# Convertir columna Gender usando laber encoder
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
x["Gender"] = encoder.fit_transform(x.values[:, 2])
x.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
sueldos = x.values[:, 9]

# EN este caso fit_transform recibe una matriz, por lo que se convierten los sueldos
# a una matriz del largo del conjunto de datos
# (-1, pues contamos desde el final) y 1 columna

sueldos_matriz = sueldos.reshape(-1, 1)
sueldos_escalados = scaler.fit_transform(sueldos_matriz)
print(sueldos_escalados)
```

```python
x["EstimatedSalary"] = sueldos_escalados
x.head()
```
