```python
# Antes que nada, importamos las librerias y recursos necesarios
## Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
#from google.colab import drive
#drive_path:str = "/content/MyDrive/"
#drive.mount(drive_path, force_remount=True)
## Declaración de LabelEncoder y MinMaxScaler para futuras operaciones
encoder = LabelEncoder()
scaler = MinMaxScaler()

## Recursos
#df = pd.read_csv(drive_path + "MyDrive/MineriaDeDatos/Evaluacion1/weatherAUS.csv") # Cambiar a drive cuando se suba a collab
df = pd.read_csv("./weatherAUS.csv") # Cambiar a drive cuando se suba a collab

df.head()
```

```python
# Obtención de la cantidad de datos nulos por columna
df.isnull().sum()
```

```python
dffordwardfill = df.groupby("Location").ffill()
dffordwardfill.isnull().sum()
```

```python
# Se reinserta la columna Location en el dataframe tratado en segunda posición
dffordwardfill['Location'] = df['Location']
# Se reorganizan las columnas para que Location quede en segunda posición
dffordwardfill = dffordwardfill.reindex(columns=['Date','Location','MinTemp', 'MaxTemp', 'Rainfall','WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm','WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm','Temp9am','Temp3pm', 'RainToday', 'RISK_MM', 'RainTomorrow'])
dffordwardfill.isnull().sum()
```

```python
# Se dropean las filas que quedaron con valores nulos
dfNoNulos = dffordwardfill.dropna()
dfNoNulos.isnull().sum()
# Se perdieron 7115 registros en el procedimiento
# Sin embargo, si hubiesos hecho esto solamente aplicando la funcion .dropna() se hubieran perdido 85773 registros,
# quedando con un total de 56420 registros.
# (menos de la mitad de los registros originales)
```

```python
#Primero solo se muestran las ciudades que tienen nombres unicos unicas
ciudades= df['Location'].unique()
#Definimos los arreglos que guardaran los datos que usaremos para la tabla que serian ciudad, temperatura maxima de la ciudad y la minima
ciud = []
tem_max_ciud = []
tem_min_ciud = []
#Iniciamos con un for con la cantidad total de ciudades con nombre unico
for i in range(len(ciudades)):
  #Filtramos los datos por ciudad
  ciud_alqui = df[df['Location'] == ciudades[i]]
  #seleccionamos los datos que necesitamos
  temper_ciud_alqui = ciud_alqui['MaxTemp']
  temper_ciud_alqui2 = ciud_alqui['MinTemp']
  #gracias a la funcion np.amax y np.amix podemos encontrar el valor maximo y minimo registrado en esa ciudad
  temper_max_ciud_alqui = np.amax(temper_ciud_alqui)
  temper_min_ciud_alqui = np.amin(temper_ciud_alqui2)
  #Guardamos los datos
  ciud.append(ciudades[i])
  tem_max_ciud.append(temper_max_ciud_alqui)
  tem_min_ciud.append(temper_min_ciud_alqui)

#Finalmente creamos el grafico
fig, ax = plt.subplots(figsize=(20, 5), dpi=100)

ax.bar(ciud , tem_max_ciud , color = 'tab:orange', label='Temperatura Máxima')

ax.bar(ciud , tem_min_ciud ,label='Temperatura Minima')
for i in range(len(ciud)):
    plt.text(ciud[i], tem_max_ciud[i] + 1, str(tem_max_ciud[i]), ha='center')

for i in range(len(ciud)):
    plt.text(ciud[i], tem_min_ciud[i] - 2, str(tem_min_ciud[i]), ha='center')
plt.xticks(rotation=90)

fig.tight_layout()
plt.xlabel('Ciudades')
plt.ylabel('Temperatura')
plt.title('Temperaturas Mínimas y Máximas por Ciudad registradas')
plt.show()
```

```python
# Transformar datos de Location, WindGustDir,WindDir9am, WindDir3pm, RainToday y RainTomorrow a valores numericos usando encoder.fit_transform
dfEncoder = dfNoNulos.copy()
dfEncoder.loc[:,'Location'] = encoder.fit_transform(dfEncoder['Location'])
dfEncoder.loc[:,'WindGustDir'] = encoder.fit_transform(dfEncoder['WindGustDir'])
dfEncoder.loc[:,'WindDir9am'] = encoder.fit_transform(dfEncoder['WindDir9am'])
dfEncoder.loc[:,'WindDir3pm'] = encoder.fit_transform(dfEncoder['WindDir3pm'])
dfEncoder.loc[:,'RainToday'] = encoder.fit_transform(dfEncoder['RainToday'])
dfEncoder.loc[:,'RainTomorrow'] = encoder.fit_transform(dfEncoder['RainTomorrow'])

dfEncoder
```

```python
# Generar matriz de correlación

dfMatrizCorrelacion = dfEncoder.loc[:,:]

# Modificamos la columna Date para solamente obtener el mes
dfMatrizCorrelacion['Date'] = pd.to_datetime(dfMatrizCorrelacion['Date'])
dfMatrizCorrelacion['Date'] = dfMatrizCorrelacion['Date'].dt.month
dfMatrizCorrelacion.rename(columns={'Date':'Month'}, inplace=True)

dfMatrizCorrelacion
```

```python
# Generar matriz de correlación
correlationMatrix = dfMatrizCorrelacion.corr()
correlationMatrix
```

```python
# Graficar matriz de correlación
plt.figure(figsize=(20, 10))
plt.matshow(correlationMatrix, fignum=1)
plt.xticks(range(len(correlationMatrix.columns)), correlationMatrix.columns, rotation=90)
plt.yticks(range(len(correlationMatrix.columns)), correlationMatrix.columns)
plt.colorbar()
plt.show()
```
