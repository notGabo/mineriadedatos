```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pltx
import plotly.express as px
```

```python
# cargamos los datos
url_base = "https://dmontero.org/mineria/datasets/StudentsPerformance.csv"
df = pd.read_csv(url_base)
df.head()
```

```python
df.info
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['user'])
```

```python
sns.pairplot(df,hue='gender')
```

```python
X = np.array(df[['math score', 'reading score', 'writing score']])
px.scatter_3d(x=X[:,0], y=X[:,1], z=X[:,2], color = df['gender'])
```

Apliquemos el algoritmo de KMeans, recuerden que kmeans me forma el numero de cluster que yo le indique

```python
from sklearn.cluster import KMeans
modelo = KMeans(n_clusters=5)
modelo.fit(X)
```

```python
etiquetas = modelo.predict(X)
px.scatter_3d(x=X[:,0], y=X[:,1], z=X[:,2], color = etiquetas)
```
